```python
# sort txt files by a new class order
###################################################
#  File path class.names
#            txt_sort.py
#               |- train,test,valid
##################################################
import os
#import csv
import pandas as pd
#pip install pandas
```

```python
# get a new class order by .names file
cl = []
names_file=[]
```

```python
for i in os.listdir('./'):
    if i.endswith('.names'):
        i = './' + i
        names_file.append(i)
if len(names_file) != 1:
    print('There are many .names file in directory!')
    names_file=[]
```

```python
print('There are defined classes below in'+ str(names_file[0])+'\n')
for item in names_file:
    name_data = []
    with open(item,'r') as names:
        for line in names:
            currentLine = line[:]
            print(currentLine)
            data = currentLine.split('\n')
            name_data.append(data)
```

```python
if x == '2' :
    del_cl = input('What class do you want to delete? : ')
    i = 0
    for k in range(len(name_data)):
        if del_cl == name_data[k][0]:
                num_del_cl = i
        i += 1
    print('Target : row number '+ str(num_del_cl))
```

```python
yy = input('Do you really want to modify '+ str(names_file[0])+' ? : (y/n)')
if yy == 'y'or yy=='yes':
    f = open(names_file[0],'w')
    ign_del_cl_num = []
    
    for i in range(len(name_data)):
        ign_del_cl_num.append(i)
    del ign_del_cl_num[num_del_cl]
    
    for i in ign_del_cl_num:
        for res in name_data[i]:
            f.write(str(res))
        f.write('\n')
    f.close()
```

```python
def make_txt(txt_file,save_path):
    for i in range(len(txt_file)):
        file = pd.read_csv(txt_file[i], sep=' ',encoding='utf-8',names=('cl','xmin','ymin','w','h'))
        #print(file)
        file_len = len(file.index)
        k = 0
        n = 0
        while n != file_len: # k column number
            if file.iloc[k][0] == num_del_cl:                    
                file.drop(k,inplace=True)
                file.reset_index(drop=True,inplace=True)
            elif file.iloc[k][0] > num_del_cl:
                file.loc[k,'cl'] = file.loc[k,'cl'] - 1
                k = k + 1
            else:
                k = k + 1
            n = n + 1
        txt_file[i] = txt_file[i][1:]
        file.to_csv('./new_txt'+txt_file[i],sep=' ',index=False, header=False) #최종경로 변경
```
